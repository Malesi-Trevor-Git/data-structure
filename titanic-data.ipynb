```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('titanic.csv')
df.replace('?',np.nan,inplace=True)
df=df.astype({"age":np.float64,"fare":np.float64})
fig,axis=plt.subplots(ncols=5,figsize=(30,5))
sns.violinplot(x="survived",y="age",hue="sex",data=df,ax=axis[0])
sns.pointplot(x="sibsp",y="survived",hue="sex",data=df,ax=axis[1])
sns.pointplot(x="parch",y="survived",hue="sex",data=df,ax=axis[2])
sns.pointplot(x="pclass",y="survived",hue="sex",data=df,ax=axis[3])
sns.violinplot(x="survived",y="fare",hue="sex",data=df,ax=axis[4])
```
